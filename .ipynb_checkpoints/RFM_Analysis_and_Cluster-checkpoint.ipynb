```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt

#sklearn
import sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv('Data/OnlineRetail.csv', encoding = "ISO-8859-1")
```

```python
df.tail()
```

```python
df.shape
```

```python
df.dropna()
```

```python
df['Amount'] = df['Quantity'] * df['UnitPrice']
monetary = df.groupby('CustomerID')['Amount'].sum()
monetary = monetary .reset_index()
monetary .tail()
```

```python
df.info()
```

```python
df.describe()
```

```python
frequency = df.groupby('CustomerID')['InvoiceNo'].count()
frequency = frequency.reset_index()
frequency.head()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], format ='%m/%d/%Y %H:%M')
df['Diff'] = max(df['InvoiceDate']) - df['InvoiceDate']
Recency = df.groupby('CustomerID')['Diff'].min()

Recency = Recency.reset_index()
Recency.head()
```

```python
Recency
```

```python
Recency['Diff'] = Recency['Diff'].dt.days
Recency.head()
```

```python
# calculating monetary by summing up all the amounts of the customer.
rfm = pd.merge(Recency, frequency, on='CustomerID', how='inner')
```

```python
rfm = pd.merge(rfm, monetary , on='CustomerID', how='inner')
```

```python
rfm.columns = ['CustomerID', 'Recency','Frequency', 'Monetory']
```

```python
rfm
```

```python
#OUTLIERS DETECT
sns.boxplot(rfm[['Recency','Frequency', 'Monetory']])
```

```python
#Remove Outliers
scaler = StandardScaler()
rfm_norm = rfm[['Recency','Frequency', 'Monetory']]
rfm_norm
```

```python
rfm_norm = scaler.fit_transform(rfm_norm)
rfm_norm = pd.DataFrame(rfm_norm)
rfm_norm
```

```python
#CLUSTERING

kmeans = KMeans(n_clusters=4, max_iter =50)
kmeans.fit(rfm_norm)
```

```python
ssd = []
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]
for num_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=num_clusters, max_iter = 50)
    kmeans.fit(rfm_norm)
    
    ssd.append(kmeans.inertia_)
```

```python
# Silhoutte Analysis
#silhoutte_score = (p-q)/max(p,q)

for num_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=num_clusters, max_iter = 50)
    kmeans.fit(rfm_norm)
    
    cluster_labels = kmeans.labels_
    
    silhouette_avg = silhouette_score(rfm_norm, cluster_labels)
    print("For n_clusters = {0}, the sil score is {1}". format(num_clusters, silhouette_avg))
```

```python
kmeans = KMeans(n_clusters=3, max_iter =50)
kmeans.fit(rfm_norm)
```

```python
rfm_norm.loc[:,'CustomerID'] = rfm['CustomerID']
rfm_norm
```

```python
rfm_norm['cluster'] = kmeans.labels_
rfm_norm
```

```python
rfm_norm.keys()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
plt.rcParams["figure.figsize"] = (20,3)
fig = plt.figure(1)
plt.clf()

ax = Axes3D(fig,
           rect = [0, 0, .95, 1],
           elev =35,
           azim = 148)

plt.cla()
ax.scatter(rfm_norm[0], rfm_norm[1], rfm_norm[2],
          c = rfm_norm['cluster'],
          s = 50,
          cmap = 'spring',
          alpha = 0.5,
          edgecolor = 'darkgrey')

ax.set_xlabel('Recency',fontsize = 10)
ax.set_ylabel('Frequency',fontsize = 10)
ax.set_zlabel('Monetary',fontsize = 10)

plt.show()
```
